```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
import geoplot as gplt
```

```python
zoneurl = 'https://data.cityofnewyork.us/resource/755u-8jsi.json'
taxizones = requests.get(zoneurl).json()
```
